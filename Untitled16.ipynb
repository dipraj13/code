```python
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
import pandas as pd

df = pd.read_csv("BostonHousing.csv")

df
```

```python
X = df.drop("MEDV", axis = 1)
Y = df["MEDV"]

X
```

```python
scalar = StandardScaler()

x_scale = scalar.fit_transform(X)
x_scale
```

```python
x_train, x_test, y_train, y_test = train_test_split(x_scale,Y, random_state=22, test_size=0.2)
x_train
```

```python
model = Sequential([
    Dense(64, input_dim = x_train.shape[1],activation = "relu"),
    Dense(32, activation = "relu"),
    Dense(1, activation="linear")
])
```

```python
model.compile(optimizer="adam", loss = "mse")

res = model.fit(x_train,y_train, epochs = 100,batch_size=32,validation_data=(x_test,y_test))
```

```python
loss = model.evaluate(x_test, y_test)
loss
```

```python
y_pred = model.predict(x_test)
y_pred
```

```python
print("\nPredictions vs Actual:")
for i in range(10):
    print(f"Prediction: {y_pred[i][0]:.2f}, Actual: {y_test.values[i]:.2f}")
```

```python
ans = pd.DataFrame({
    "Actual" : y_test.values,
    "Prediction" : y_pred.flatten()
})
ans.to_csv("ans.csv", index=False)
```

```python
import matplotlib.pyplot as plt

plt.plot(res.history["loss"], label = "Training Loss")
plt.plot(res.history["val_loss"],label = "Validation Loss")
plt.title("Training and Validation Loss")
plt.xlabel("epochs")
plt.ylabel("loss")
plt.legend()
plt.show()
```
